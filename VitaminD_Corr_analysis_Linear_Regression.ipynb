```python
# оцiнка параметра β методом 1МНК (b = β*x параметри моделi, якi необхiдно оцiнити)

def getB(x,y):
    return pd.DataFrame(np.dot(np.dot(np.linalg.inv(np.dot(x.T,x)), x.T), y))
```

```python
def y(x,b):
    return np.dot(x, b)
```

# 4. Показники адекватностi регресiйної моделi

```python
# Коефiцiєнт детермiнацiї R^2

def R2(y1,yt):
    ym = y1.mean()
    return (((yt-ym)*(yt-ym)).sum()) / (((y1-ym)*(y1-ym)).sum())
```

```python
# F-статистика - для визначення значущостi регресiйного рiвняння

def Fst(r2,t,n):
    return r2*(t-n) / ((1-r2*r2)*(n-1))
```

```python
# F-критерій Фішера
import scipy.stats as stats
def Fkr(alpha, t, n):
  stats.f.ppf(q=1-alpha, dfn= n-1, dfd= t - n)
```

```python
# Коефiцiєнт детермiнацiї, скоригований за Тейлом

def R2T(r2, t, n):
    return 1 - (1 - r2) * (t - 1) / (t - n)
```

```python
# Коефiцiєнт детермiнацiї, скоригований за Амемiєю

def R2A(r2, t, n):
    return 1 - (1 - r2) * (t + n) / (t - n)   
```

```python
# t-критерий 
from scipy import stats
def t(alpha, gl):
    return stats.t.ppf(1 - (alpha / 2), gl)

print(t(0.05, 928-2))
```

## Частковий коефiцiєнт детермiнацiї

```python
def s2(y1, yt, n):
    return ((y1 - yt) * (y1 - yt)).sum() / (T - n)

def s2Bk(x, y1, yt, n):
    return pd.DataFrame(np.linalg.inv(np.dot(x.T, x))) * (s2(y1, yt, n)[0])

def dR2k(r2, t, n, b, s2bk, k):
    return (1 - r2) * ((b[k] / s2bk[k][k])**2) / (t - n)
```
